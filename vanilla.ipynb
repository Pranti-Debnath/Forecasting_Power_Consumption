```python
import shap
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
# Preprocessing: Convert Datetime to a usable format and extract features
small_data['Datetime'] = pd.to_datetime(small_data['Datetime'], format='%m/%d/%Y %H:%M')
small_data['Hour'] = small_data['Datetime'].dt.hour
small_data['Day'] = small_data['Datetime'].dt.day
small_data['Month'] = small_data['Datetime'].dt.month
small_data['Weekday'] = small_data['Datetime'].dt.weekday
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(train_data.iloc[:, 1:])  # Exclude 'Datetime'

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length, :])  # All input features
        y.append(data[i + seq_length, -1])  # Only the target variable
    return np.array(X), np.array(y)



seq_length = 96  # Number of past time steps to consider
X_train, y_train = create_sequences(scaled_data, seq_length)

# Repeat for validation and test sets
scaled_val = scaler.transform(val_data.iloc[:, 1:])
scaled_test = scaler.transform(test_data.iloc[:, 1:])
X_val, y_val = create_sequences(scaled_val, seq_length)
X_test, y_test = create_sequences(scaled_test, seq_length)

# Print the shapes of tokenized data
print(f"Training data shape: {X_train.shape}, {y_train.shape}")
print(f"Validation data shape: {X_val.shape}, {y_val.shape}")
print(f"Test data shape: {X_test.shape}, {y_test.shape}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import math

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)  # Shape: (1, max_len, d_model)
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]
class VanillaTransformer(nn.Module):
    def __init__(self, input_dim, d_model, nhead, num_layers, seq_length, dropout=0.1):
        super(VanillaTransformer, self).__init__()
        self.input_embedding = nn.Linear(input_dim, d_model)
        self.positional_encoding = nn.Parameter(
            torch.zeros(seq_length, d_model)
        )  # Learnable positional encoding for time series
        self.transformer_encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, dim_feedforward=4*d_model, dropout=dropout),
            num_layers=num_layers
        )
        self.layer_norm = nn.LayerNorm(d_model)  # Layer normalization
        self.fc_out = nn.Linear(d_model, 1)  # Output layer for regression

    def forward(self, x):
        x = self.input_embedding(x) + self.positional_encoding
        x = self.transformer_encoder(x)
        x = self.layer_norm(x)  # Normalize output from Transformer
        return self.fc_out(x[:, -1, :])  # Use the last output for prediction


    def forward(self, x):
        x = self.input_embedding(x) + self.positional_encoding
        x = self.transformer_encoder(x)
        x = self.layer_norm(x)  # Normalize output from Transformer
        return self.fc_out(x[:, -1, :])  # Use the last output for prediction


    def forward(self, x):
        x = self.input_embedding(x) + self.positional_encoding
        x = self.transformer_encoder(x)
        x = self.layer_norm(x)  # Normalize output from Transformer
        return self.fc_out(x[:, -1, :])  # Use the last output for prediction

# Hyperparameters
input_dim = X_train.shape[2]
seq_length = X_train.shape[1]
d_model = 64
nhead = 4
num_layers = 3
dropout = 0.05

# Initialize model
model = VanillaTransformer(input_dim, d_model, nhead, num_layers, seq_length, dropout)

# Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Loss function & optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
# Add learning rate scheduler
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)  # Reduce LR by 50% every 5 epochs

# Add learning rate scheduler
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)  # Reduce LR by 50% every 5 epochs


# Convert data to PyTorch tensors
train_tensor = torch.tensor(X_train, dtype=torch.float32).to(device)
target_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(-1).to(device)

# Training loop
epochs = 10
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(train_tensor)
    loss = criterion(output, target_tensor)
    loss.backward()
    optimizer.step()
    scheduler.step()  # Step the scheduler at the end of each epoch
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss.item()}")



# Model evaluation on test set
model.eval()
with torch.no_grad():
    test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
    y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(-1).to(device)
    predictions = model(test_tensor).cpu().numpy()
    true_values = y_test_tensor.cpu().numpy()

# Compute evaluation metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(true_values, predictions)
rmse = mse ** 0.5
mae = mean_absolute_error(true_values, predictions)
r2 = r2_score(true_values, predictions)

print(f"Evaluation Metrics:\nMSE: {mse:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")
```
